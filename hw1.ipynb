```python
#I need to pick a small range
file=open("./cc1.din.txt")
line=file.readline()
address_dict={}
while line:
    line=line.strip("\n")
    address_name=line.split(" ")[1]
    address_name=int(address_name,16)
    #print(address_name)
    if address_name<4231868 and address_name>4230868:
        if address_name in address_dict:
            address_dict[address_name]+=1
        else:
             address_dict[address_name]=1
    line=file.readline()
x=list(address_dict.keys())
y=list(address_dict.values())
plt.bar(x,y)
```

```python
#count the frequency of read and write
file=open("./cc1.din.txt")
line=file.readline()
address_dict={"read":0,"write":0,"instruction fetch":0}
while line:
    line=line.strip("\n")
    operation_name=line.split(" ")[0]
    if operation_name=="0":
        address_dict["read"]+=1
    if operation_name=="1":
        address_dict["write"]+=1
    if operation_name=="2":
        address_dict["instruction fetch"]+=1
    line=file.readline()
```

```python
def int_matrix_produce(n,m):
    matrix_1=np.random.randint(0,99,size=(n,m))
    matrix_2=np.random.randint(0,99,size=(m,n))
    return matrix_1,matrix_2
```

```python
def double_matrix_produce(n,m):
    matrix_1=np.random.uniform(0,99,size=(n,m))
    matrix_2=np.random.uniform(0,99,size=(m,n))
    return matrix_1,matrix_2
```

```python
a,b=int_matrix_produce(4,5)
```

```python
a,b=double_matrix_produce(4,5)
```

```python
def matrix_multiply(matrix_1,matrix_2):
    results=np.zeros((matrix_1.shape[0],matrix_2.shape[1]))
    for i in range(matrix_1.shape[0]):
        for j in range(matrix_2.shape[1]):
            total_sum=0
            for k in range(matrix_1.shape[1]):
                total_sum+=matrix_1[i][k]*matrix_2[k][j]
            results[i][j]=total_sum
    return results
```

```python
matrix_multiply(a,b)
```

```python
matrix_multiply(a,b)
```
